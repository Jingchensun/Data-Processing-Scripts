```python
import json
import os

FILE_PATH = "/home/jingchen/data/natural-instructions/natural-instructions-2.8/tasks/"
# FILE_PATH = "/home/jingchen/data/natural-instructions-2.8/test/"

# get all the images under the file
image_list = sorted(os.listdir(FILE_PATH))

# OUT_FILE2 = "/home/jingchen/data/natural-instructions/new_natural/Pure_English.jsonl"
OUT_FILE2 = "/home/jingchen/data/natural-instructions/new_natural/Mix_English.jsonl"
print("out file2:", OUT_FILE2)
with open(OUT_FILE2, "w") as k:
    for i in range(len(image_list)):
        FILE_NAME = FILE_PATH + image_list[i]
        print("File name:", FILE_NAME)

        with open(FILE_NAME, "r") as f:
            data = json.load(f)
            input_num = len(data["Input_language"])
            output_num = len(data["Output_language"])
            I1 = data["Input_language"][0]
            O1 = data["Output_language"][0]

        # if input_num==output_num==1:
    
        if I1 == O1 == "English":
            print(I1,O1)
            print("yesyesyesyesyesyesyes")

            task_definition = " ".join(data["Definition"])
            # OUT_PATH = "/home/jingchen/data/natural-instructions/new_natural/Pure_English/"
            # OUT_FILE1 = OUT_PATH + image_list[i]
            # print("out file1:", OUT_FILE1)
            # OUT_FILE2 = str(OUT_FILE1).replace("json", "jsonl")
            for instance in data["Instances"]:

                # Create the "src" by combining task definition and input
                src = task_definition + " " + instance["input"]

                # Concatenate all outputs for "trg"
                trg = " ".join(instance["output"])

                # Create a dictionary with "src" and "trg"
                result = {"src": src, "trg": trg}

                # Write each dictionary as a single line in the output file
                k.write(json.dumps(result) + "\n")
```
