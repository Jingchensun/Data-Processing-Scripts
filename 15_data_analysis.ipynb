```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
data = pd.read_excel('./input/ablition.xlsx')

# 获取数据集名称列表
datasets = data.columns[2:-1]

# 获取不同的 lamda1-lamda2 列表
lamda_values = data['lamda1-lamda2'].unique()

# 定义颜色列表，确保足够的颜色以区分不同的 lamda1-lamda2
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# 设置柱状图的宽度
bar_width = 0.15

# 绘制柱状图
plt.figure(figsize=(12, 8))

# 绘制不同的 lamda1 和 lamda2
for i, (lamda, color) in enumerate(zip(lamda_values, colors)):
    lamda_data = data[data['lamda1-lamda2'] == lamda]
    bar_positions = np.arange(len(datasets)) + i * bar_width
    # plt.bar(bar_positions, lamda_data.iloc[0, 2:-1], width=bar_width, label=f"lamda1-lamda2={lamda}", color=color)
    # plt.plot(datasets, lamda_data.iloc[0, 2:-1], label=f"$\lambda_1$-$\lambda_2$={lamda}", color=color, marker='o', linestyle='-', linewidth=3)
    plt.bar(bar_positions, lamda_data.iloc[0, 2:-1], width=bar_width, label=f"$\lambda_1$-$\lambda_2$={lamda}", color=color)

# 设置图表标题和标签
# plt.title('Accuracy vs. Datasets for Different lamda1-lamda2', fontsize=16)
# plt.xlabel('Datasets', fontsize=14)
plt.ylabel('Accuracy', fontsize=20)

# 调整图例字体大小
plt.legend(fontsize=20)

# 旋转横坐标标签，以免重叠
plt.xticks(np.arange(len(datasets)) + bar_width * (len(lamda_values) - 1) / 2, datasets, rotation=45, ha='right', fontsize=20)

# 增大刻度标签字体大小
plt.yticks(fontsize=20)

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_excel('./input/unicl.xlsx')

# 设置图表样式
plt.style.use('ggplot')

# 提取方法名称和精度数据
methods = data['Method']
accuracy_data = data.iloc[:, 1:]

# 绘制柱状图
plt.figure(figsize=(12, 8))
for i, method in enumerate(methods):
    plt.bar(range(len(accuracy_data.columns)), accuracy_data.iloc[i, :], label=method, alpha=0.7)

# 设置图表标题和标签
plt.title('Accuracy Comparison Across Methods and Datasets', fontsize=16)
plt.xlabel('Datasets', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)

# 设置横坐标刻度标签
plt.xticks(range(len(accuracy_data.columns)), accuracy_data.columns, rotation=45, ha='right', fontsize=12)

# 设置图例
plt.legend(fontsize=12)

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_excel('./input/unicl.xlsx')

# 获取数据集名称列表
datasets = data.columns[2:-1]

# 获取不同的 lamda1-lamda2 列表
lamda_values = data['Method'].unique()

# 定义颜色列表，确保足够的颜色以区分不同的 lamda1-lamda2
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# # 使用seaborn的风格
# plt.style.use('seaborn-darkgrid')

# 绘制折线图
plt.figure(figsize=(12, 8))

# 定义不同方法的颜色、样式、符号等
line_styles = ['-', '--', '-.', ':', '-', '--', '-.']
markers = ['o', 's', '^', 'v', 'D', 'p', 'H']

# 绘制不同的 lamda1 和 lamda2
for i, (lamda, color, linestyle, marker) in enumerate(zip(lamda_values, colors, line_styles, markers)):
    lamda_data = data[data['Method'] == lamda]
    plt.plot(datasets, lamda_data.iloc[0, 2:-1], label=f"{lamda}", color=color, marker=marker, linestyle=linestyle, linewidth=3, markersize=8)

# 设置图表标题和标签
# plt.title('Accuracy vs. Datasets for Different lamda1-lamda2', fontsize=16)
plt.xlabel('Datasets', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)

# 设置横坐标刻度标签
plt.xticks(rotation=45, ha='right', fontsize=20)

# 调整图例字体大小
plt.legend(fontsize=18)

# 显示图表
plt.tight_layout()
plt.show()
```
