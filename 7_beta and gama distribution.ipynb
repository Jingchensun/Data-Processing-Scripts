```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 创建一个Beta分布对象
beta_dist = beta(0.5, 0.5)

# 生成一系列x值，用于绘制PDF
x = np.linspace(0, 1, 1000)

# 计算Beta分布的PDF
pdf = beta_dist.pdf(x)

# 创建PDF图形
plt.plot(x, pdf, 'b-', lw=2, label='PDF')
plt.title('Beta (PDF)')
plt.xlabel('x')
plt.ylabel("pd")
plt.legend()

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 创建一系列x值，用于绘制贝塔分布的图形
x = np.linspace(0, 1, 1000)

# 不同参数组合的贝塔分布
alpha_values = [0.3, 0.5, 0.7, 1, 2, 3, 5]
beta_values = [0.3, 0.5, 0.7, 1, 2, 3, 5]

# 绘制不同参数组合的贝塔分布
plt.figure(figsize=(12, 8))
for alpha, b in zip(alpha_values, beta_values):
    # 创建贝塔分布对象
    print(alpha, b)
    beta_dist = beta(alpha, b)
    
    # 计算贝塔分布的概率密度函数（PDF）
    pdf = beta_dist.pdf(x)
    
    # 绘制贝塔分布的图形
    plt.plot(x, pdf, lw=2, label=f'alpha={alpha}, beta={b}')

# plt.title('不同参数取值情况下的贝塔分布')
# plt.xlabel('随机变量值 (x)')
# plt.ylabel('概率密度')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# 设置Gamma分布的参数
alpha = 2.0  # 形状参数
beta = 1.0   # 尺度参数

# 生成一组随机样本
data = gamma.rvs(alpha, scale=beta, size=1000)

# 生成一组x值，用于绘制概率密度函数
x = np.linspace(0, 10, 100)
pdf = gamma.pdf(x, alpha, scale=beta)

# 绘制Gamma分布的概率密度函数
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, 'r-', lw=2, label='Gamma PDF')
plt.hist(data, bins=20, density=True, alpha=0.5, color='blue', label='Sample Data')
plt.title('Gamma Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 设置Beta分布的参数
alpha = 2.0  # 第一个形状参数
beta_param = 5.0  # 第二个形状参数

# 生成一组随机样本
data = beta.rvs(alpha, beta_param, size=1000)

# 生成一组x值，用于绘制概率密度函数
x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, alpha, beta_param)

# 绘制Beta分布的概率密度函数
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, 'r-', lw=2, label='Beta PDF')
plt.hist(data, bins=20, density=True, alpha=0.5, color='blue', label='Sample Data')
plt.title('Beta Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# 不同的Gamma分布参数
alpha_values = [0.5, 2, 4, 4]
beta_values = [1, 1, 2, 6]

# 生成一组x值，用于绘制概率密度函数
x = np.linspace(0, 10, 100)

# 绘制不同参数下的Gamma分布的概率密度函数
plt.figure(figsize=(12, 8))
for alpha, beta in zip(alpha_values, beta_values):
    pdf = gamma.pdf(x, alpha, scale=1/beta)  # 注意：SciPy中的scale参数是1/β
    plt.plot(x, pdf, lw=2, label=f'alpha={alpha}, beta={beta}')

plt.title('Gamma Distribution with Different Parameters')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
