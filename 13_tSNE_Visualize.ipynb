```python
import torch
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
num_samples = 50
num_features = 512  # CLIP模型的特征维度

# 使用均值为0，标准差为1的正态分布生成随机数据
random_data = np.random.randn(num_samples, num_features)

# 使用t-SNE进行降维
tsne = TSNE(n_components=3, random_state=42)
embeddings = tsne.fit_transform(random_data)

# 可视化
plt.scatter(embeddings[:, 0], embeddings[:, 1])
plt.title('t-SNE Visualization of Random Data')
plt.show()
```
