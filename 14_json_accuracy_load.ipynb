```python
import os
import json
import numpy as np

def calculate_mean_and_variance(json_data):
    # 获取最后一个{}中的"origin_acc0", "origin_acc1", "origin_acc2"对应的值
    last_entry = json_data
    origin_acc_values = [last_entry.get("origin_acc0", 0),
                         last_entry.get("origin_acc1", 0),
                         last_entry.get("origin_acc2", 0)]
    
    mean_value = np.mean(origin_acc_values)
    print("mean_value:", mean_value)
    variance_value = np.var(origin_acc_values)
    return mean_value, variance_value

def process_json_file(file_path):
    with open(file_path, 'r') as file:
        try:
            data = json.load(file)
            print(data)
            mean, variance = calculate_mean_and_variance(data)
            print(f"File: {file_path}")
            print(f"Mean: {mean}")
            print(f"Variance: {variance}")

            # 将计算得到的variance_value写入到json文件中
            last_entry = data
            last_entry["variance"] = variance

            # 写回到文件
            with open(file_path, 'w') as output_file:
                json.dump(data, output_file, indent=4)

            print("Variance written back to file.")
            print("----------")
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON in file {file_path}: {e}")
        except (KeyError, IndexError) as e:
            print(f"Error in file {file_path}: {e}")

def main():
    # 遍历output文件夹下的所有json文件
    output_folder = "/home/jingchen/tip-adapter/output"  # 替换为你的实际文件夹路径
    for file_name in os.listdir(output_folder):
        if file_name.endswith(".json"):
            file_path = os.path.join(output_folder, file_name)
            print(file_path)
            process_json_file(file_path)

if __name__ == "__main__":
    main()
```
