```python
import os  
import cv2  
from PIL import Image 

def image_to_video(image_path, media_path):
    '''
    :param image_path: images path
    :param media_path: path to save video
    :return:
    '''
    # get all the images under the file
    image_names = os.listdir(image_path)
    # sort the images with the name
    image_names.sort(key=lambda n: int(n[:-4]))
    # set the write format of video
    fourcc = cv2.VideoWriter_fourcc('M', 'P', '4', 'V')
    # set the frame rate
    fps = 3  # set the frame rate = 3
    # get the image size with first picture
    image = Image.open(image_path + image_names[0])
    # initilize the wirte object
    media_writer = cv2.VideoWriter(media_path, fourcc, fps, image.size)
    # laod every images
    for image_name in image_names:
        im = cv2.imread(os.path.join(image_path, image_name))
        media_writer.write(im)
        print(image_name, 'Merged!')
    # Release
    media_writer.release()
    print('Finish writing!')
file = "test"
video_name = file + "_merge.mp4"
image_to_video("output/" + file + "/origin/", "output/"  + file +"/" + video_name ) # 
```
