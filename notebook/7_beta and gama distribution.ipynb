```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 创建一个Beta分布对象
beta_dist = beta(0.5, 0.5)

# 生成一系列x值，用于绘制PDF
x = np.linspace(0, 1, 1000)

# 计算Beta分布的PDF
pdf = beta_dist.pdf(x)

# 创建PDF图形
plt.plot(x, pdf, 'b-', lw=2, label='PDF')
plt.title('Beta (PDF)')
plt.xlabel('x')
plt.ylabel("pd")
plt.legend()

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# 设置Gamma分布的参数
alpha = 2.0  # 形状参数
beta = 1.0   # 尺度参数

# 生成一组随机样本
data = gamma.rvs(alpha, scale=beta, size=1000)

# 生成一组x值，用于绘制概率密度函数
x = np.linspace(0, 10, 100)
pdf = gamma.pdf(x, alpha, scale=beta)

# 绘制Gamma分布的概率密度函数
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, 'r-', lw=2, label='Gamma PDF')
plt.hist(data, bins=20, density=True, alpha=0.5, color='blue', label='Sample Data')
plt.title('Gamma Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 定义泊松分布的参数 lambda（平均发生率）
lambda_param = 3.0

# 创建一个泊松分布随机变量对象
poisson_dist = poisson(mu=lambda_param)

# 生成随机样本（事件发生的次数）
random_samples = poisson_dist.rvs(size=1000)

# 绘制直方图来表示随机样本的分布
plt.hist(random_samples, bins=20, density=True, alpha=0.6, color='g', label='Sampled Data')

# 计算并绘制泊松分布的概率质量函数（PMF）
x = np.arange(0, 15)
pmf_values = poisson_dist.pmf(x)
plt.plot(x, pmf_values, 'ro-', label='Poisson PMF')

plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm, kde

# 定义泊松分布的参数 lambda（平均发生率）
lambda_param = 3.0

# 创建一个泊松分布随机变量对象
poisson_dist = poisson(mu=lambda_param)

# 生成随机样本（事件发生的次数）
random_samples = poisson_dist.rvs(size=1000)

# 使用核密度估计来获得连续曲线
kde_estimator = kde.gaussian_kde(random_samples)
x_values = np.linspace(0, 15, 100)
kde_values = kde_estimator(x_values)

# 绘制核密度估计的连续曲线
plt.plot(x_values, kde_values, 'r-', label='KDE')

# 绘制泊松分布的概率质量函数（PMF）作为比较
x = np.arange(0, 15)
pmf_values = poisson_dist.pmf(x)
plt.plot(x, pmf_values, 'bo-', label='Poisson PMF')

plt.xlabel('Number of Events')
plt.ylabel('Probability Density')
plt.title('Poisson Distribution vs. KDE')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 不同的 alpha 值
alphas = [1.0, 2.0, 3.0]
x_m = 1.0  # 固定最小值参数

# 生成不同参数下的 Pareto 分布 PDF
x = np.linspace(1, 10, 1000)  # 生成 x 范围
plt.figure(figsize=(10, 6))

for alpha in alphas:
    pdf = (alpha * (x_m ** alpha)) / (x ** (alpha + 1))
    plt.plot(x, pdf, label=f'Alpha = {alpha}')

plt.xlabel('Value (x)')
plt.ylabel('Probability Density')
plt.title('Pareto Distribution PDF for Different Alpha Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 不同的 alpha 和 x_m 值
alphas = [1.0, 2.0, 3.0]
x_ms = [1.0, 2.0, 3.0]

# 生成不同参数下的 Pareto 分布 PDF
x = np.linspace(1, 10, 1000)  # 生成 x 范围
plt.figure(figsize=(12, 8))

for alpha in alphas:
    for x_m in x_ms:
        pdf = (alpha * (x_m ** alpha)) / (x ** (alpha + 1))
        label = f'Alpha = {alpha}, x_m = {x_m}'
        plt.plot(x, pdf, label=label)

plt.xlabel('Value (x)')
plt.ylabel('Probability Density')
plt.title('Pareto Distribution PDF for Different Alpha and x_m Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 Gumbel 分布的参数
mu = 0  # 位置参数
beta = 1  # 尺度参数

# 生成 Gumbel 随机样本
sample_size = 1000
gumbel_samples = mu - beta * np.log(-np.log(np.random.rand(sample_size)))

# 生成一组 x 值，用于绘制 PDF
x = np.linspace(-10, 10, 1000)

# 计算 Gumbel 分布的 PDF
gumbel_pdf = (1/beta) * np.exp(-(x - mu)/beta) * np.exp(-np.exp(-(x - mu)/beta))

# 创建一个图形窗口
plt.figure(figsize=(8, 5))

# 绘制 Gumbel 分布的直方图
plt.hist(gumbel_samples, bins=50, density=True, alpha=0.7, color='blue', label='Gumbel Samples', zorder=2)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gumbel Distribution')
plt.grid(True)

# 绘制 Gumbel 分布的 PDF
plt.plot(x, gumbel_pdf, color='red', label='Gumbel PDF', zorder=1)
plt.legend(loc='upper right')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 生成随机数据
x = np.linspace(-5, 5, 1000)

# 正态分布的概率密度函数 (PDF)
pdf = norm.pdf(x, loc=0, scale=1)

# 正态分布的累积分布函数 (CDF)
cdf = norm.cdf(x, loc=0, scale=1)

# 绘制PDF图像
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'r-', label='PDF')
plt.title('Probability Density Function (PDF)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# 绘制CDF图像
plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'b-', label='CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# 成功的概率
p = 0.3

# 生成伯努利分布的随机变量
rv = bernoulli(p)

# 生成取值
x = [0, 1]

# 计算PDF
pdf = [rv.pmf(val) for val in x]

# 计算CDF
cdf = [rv.cdf(val) for val in x]

# 绘制PDF图像
plt.subplot(1, 2, 1)
plt.bar(x, pdf, width=0.1, align='center', alpha=0.5)
plt.xticks(x)
plt.title('Probability Density Function (PDF)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.ylim(0, 1)

# 绘制CDF图像
plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post')
plt.xticks(x)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
