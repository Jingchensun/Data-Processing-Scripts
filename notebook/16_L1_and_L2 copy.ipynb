```python
import numpy as np
import matplotlib.pyplot as plt

# 定义向量的坐标
x = np.linspace(-5, 5, 100)  # 生成从 -5 到 5 之间的 100 个均匀分布的数作为向量的 x 坐标
y = np.linspace(-5, 5, 100)  # 生成从 -5 到 5 之间的 100 个均匀分布的数作为向量的 y 坐标

# 生成网格点坐标
X, Y = np.meshgrid(x, y)

# 计算每个点的二阶范数
Z = np.sqrt(X**2 + Y**2)

# 绘制二阶范数图像
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap='viridis')  # 绘制等高线填充图
plt.colorbar()  # 添加颜色条
plt.xlabel('X')
plt.ylabel('Y')
plt.title('二阶范数图像')
plt.grid(True)
plt.show()
```
