```python
import numpy as np
from PIL import Image
import json
import cv2
import os


def save_image(load_dict, rectangle):
    print('load_dict:', load_dict)
    img_name = load_dict['img_name']
    bbox = load_dict["bbox_anno"]
    class_name = load_dict['class_name']
    image = np.array(Image.open('input/' + img_name))
  

    x1 = int(bbox[0])
    y1 = int(bbox[1])
    x2 = int(bbox[2])
    y2 = int(bbox[3])
    x_center = int(x1 + (x2-x1)/2)
    y_center = int(y1 + (y2-y1)/2)
    w = x2 - x1
    h = y2 - y1
    s_min = min(w, h)
    s_max = max(w, h)

    #draw rectangle
    if rectangle == True:
        cv2.rectangle(image, (x1, y1), (x2, y2), (0,255,0), 4) 
        # factor_r = 3, ratio = (1 - (1/3)*2)^2 = 11.1%
        # factor_r = 4, ratio = (1 - (1/4)*2)^2 = 25.0%
        # factor_r = 5, ratio = (1 - (1/5)*2)^2 = 36.0%
        # factor_r = 6, ratio = (1 - (1/6)*2)^2 = 44.4%
        # factor_r = 7, ratio = (1 - (1/7)*2)^2 = 51.0%
        factor_r = 3 #3,4,5,6,7
        cv2.rectangle(image, (x1 + int(w/factor_r), y1 + int(h/factor_r)), (x2 - int(w/factor_r), y2 - int(h/factor_r)), (0,255,0), -1) 
        ratio = (1 - (1/factor_r)*2)
        ratio = ratio * ratio
        print('ratio:', ratio)

    #draw circle
    else:
        cv2.rectangle(image, (x1, y1), (x2, y2), (0,255,0), 4) 
        # factor_c = 2, ratio = pi/4 * (s_min/s_max) = 0.785 * (s_min/s_max)
        # factor_c = 3, ratio = pi/9 * (s_min/s_max) = 0.349 * (s_min/s_max)
        # factor_c = 4, ratio = pi/16 * (s_min/s_max) = 0.196 * (s_min/s_max)
        # factor_c = 5, ratio = pi/25 * (s_min/s_max) = 0.125 * (s_min/s_max)
        # factor_c = 6, ratio = pi/36 * (s_min/s_max) = 0.087 * (s_min/s_max)
        # factor_c = 7, ratio = pi/49 * (s_min/s_max) = 0.064 * (s_min/s_max)
        # factor_c = 8, ratio = pi/64 * (s_min/s_max) = 0.049 * (s_min/s_max)
        factor_c = 2 #2,3,4,5
        cv2.circle(image,(x_center, y_center), int(s_min/factor_c), (0,0,255),-1)  
        ratio = (3.1415926/(factor_c * factor_c)) * (s_min/s_max)
        print('ratio:', ratio)

    #draw class_name
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(image, str(class_name), (x1-3, y1-3), font, 1, (0,0,255), 2)
    # Image.fromarray(image)

    dir_subfle = img_name.split("/")[0]
    dir_imgname = img_name.split("/")[1]
    if rectangle == True:
        dir_save = "output/" + "rectangle/" + str(ratio) #+ "/" + str(dir_subfle)
    else:
        if 0 < ratio < 0.1:
            dir_save = "output/" + "circle/" + "0.10/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.1 < ratio < 0.2:
            dir_save = "output/" + "circle/" + "0.20/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.2 < ratio < 0.3:
            dir_save = "output/" + "circle/" + "0.30/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.3 < ratio < 0.4:
            dir_save = "output/" + "circle/" + "0.40/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.4 < ratio < 0.5:
            dir_save = "output/" + "circle/" + "0.50/" + str(ratio) #+ "/" + str(dir_subfle)            
        elif 0.5 < ratio < 0.6:
            dir_save = "output/" + "circle/" + "0.60/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.6 < ratio < 0.7:
            dir_save = "output/" + "circle/" + "0.70/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.7 < ratio < 0.8:
            dir_save = "output/" + "circle/" + "0.80/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.8 < ratio < 0.9:
            dir_save = "output/" + "circle/" + "0.90/" + str(ratio) #+ "/" + str(dir_subfle)
        elif 0.9 < ratio < 1.0:
            dir_save = "output/" + "circle/" + "1.00/" + str(ratio) #+ "/" + str(dir_subfle)

    if not os.path.exists(dir_save):
        os.makedirs(dir_save)
    print(dir_save)
    cv2.imwrite(dir_save + "/" + dir_imgname, image)


submit = './output/test.json'
with open(submit,'r') as load_f:
    load_dict = json.load(load_f)
    for i in range(len(load_dict)):
        save_image(load_dict[i], rectangle=True)
```
